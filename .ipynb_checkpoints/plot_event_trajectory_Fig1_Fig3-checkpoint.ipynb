```python
def plot_sequence(time_history, num_history):
    """
    time: one d sequence, n
    num: two d, L*s, (m1, n1, ...)
    """
    bcr_binding = 1
    apc_binding = 2
    bcr_unbinding = 3
    apc_unbinding = 4
    
    def determine_reac_type(a1, a2):
        """
        return the type of reaction based on 
        bond conformation
        """
        if a1[0]<a2[0] and a1[1] ==a2[1]:
            ## APC side binding
            return apc_binding
        elif a1[0]<a2[0] and a1[1]>a2[1]:
            ## BCR side binding
            return bcr_binding
        
        elif a1[0]>a2[0] and a1[1]==a2[1]:
            ## APC side unbinding
            return apc_unbinding
        elif a1[0]>a2[0] and a1[1]<a2[1]:
            ### BCR side unbinding
            return bcr_unbinding
        else:
            print("recation type error ")
            return 0
    
    
    event_times = [[], [], [], [], []]
    
    for i, ti in enumerate(time_history):
        if i ==0:
            continue
        event_times[determine_reac_type(num_history[i-1], num_history[i])].append(ti/1000)
    
    fig, ax = plt.subplots(figsize=(6,2), dpi=100)
    plt.xlabel("time, s")
    #plt.ylabel("events")
    plt.ylim(-1.2, 1.2)
    #plt.scatter(event_times[1], [1]*len(event_times[1]), s=8, alpha=0.6)
    #plt.scatter(event_times[2], [2]*len(event_times[2]), s=8, alpha=0.6)
    #plt.scatter(event_times[3], [-1]*len(event_times[3]), s=8, alpha=0.6)
    #plt.scatter(event_times[4], [-2]*len(event_times[4]), s=8, alpha=0.6)
    
    width = 0.001
    
    plt.bar(event_times[1], [1]*len(event_times[1]), width=width, alpha=1, color="C0")
    plt.bar(event_times[2], [1]*len(event_times[2]), width=width, alpha=1, color="C3")
    plt.bar(event_times[3], [-1]*len(event_times[3]), width=width, alpha=1, color="C0")
    plt.bar(event_times[4], [-1]*len(event_times[4]), width=width, alpha=1, color="C3")
    ax.axhline(y=0, xmin=0, xmax=time_history[-1], color="gray")
    ax.set_yticks([-0.5, 0.5])
    ax.set_yticklabels(["unbinding ", "  binding"],rotation='vertical')
    for tick in ax.yaxis.get_major_ticks():
        tick.label1.set_verticalalignment('center')
    plt.show()
    return event_times

et = plot_sequence(sto1.t_record, sto1.spec_record)

pass
```

```python
def get_fm_gm(f0):
    sto2.prm_dict["f0"] = f0
    l0 = sto2.prm_dict["l0_list"][0]
    
    print("l0=", l0, ", f0=", f0*l0)
    sto2.loadSharing = False
    sto2.setup()
    sto2.run1(output=False)
    fm2 = np.array([1.0*f0]*l0)
    mask2 = sto2.spec_record[1:, 1] - sto2.spec_record[:-1, 1]
    fm2[mask2==0] = np.nan
    gm2 = l0*[sto2.force.get_f(t=1)]

    sto2.prm_dict["f0"] = l0*f0
    sto2.loadSharing = True
    sto2.setup()
    sto2.run1(output=False)
    fm = sto2.force.get_f(t=1)/sto2.spec_record[:-1, 0]
    mask = sto2.spec_record[1:, 1] - sto2.spec_record[:-1, 1]
    gm = sto2.force.get_f(t=1)/sto2.spec_record[:-1, 0]
    fm[mask==0] = None
    return fm2, fm, gm2, gm


fm1, fm01, gm1, gm01 = get_fm_gm(1)

# fm2, fm02, gm2, gm02 = get_fm_gm(10)

# fm3, fm03, gm3, gm03 = get_fm_gm(40)
```
